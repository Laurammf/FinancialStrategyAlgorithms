* per i giorni in cui la colonna precedente passa da False a True, scrive il prezzo col segno negativo (simula un acquisto)
* per i giorni in cui la colonna precedente passa da True a False, scrive il prezzo col segno positivo (simula una vendita)
* gli altri giorni metti il valore 0
5. Controlla che se la nuova colonna inizia con una vendita allora finisca con un acquisto, mentre se inizia con un acquisto allora finiosca con una vendita. Se non è così, metti a 0 l'ultimo valore diverso da 0.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
import yfinance as yf
```

```python
#1. Scarica i prezzi giornalieri (adjusted close) per 10 anni di un titolo a tua scelta e inseriscili in un dataframe.
tsla=yf.download("TSLA", start="2011-10-15", end="2021-10-15")
```

```python
tsla.head(21)
```

```python
#2. Calcola la media mobile semplice a 30 giorni e inseriscila nel dataframe.
tsla.loc[:,"SMA30"]=tsla["Adj Close"].rolling(window=30, min_periods=20).mean()
```

```python
tsla.head(21)
```

```python
# Cancel the NANs
tsla=tsla.dropna()
tsla.head(21)
```

```python
#3. Crea una nuova colonna del dataframe che dice se il prezzo di oggi è maggiore alla media mobile.
tsla.loc[:,"SMABigger"]=tsla["Adj Close"]>tsla["SMA30"]
```

```python
tsla.head(21)
```

```python
tsla.loc[:,"SimpleAvgPosition"] = ""
```

```python
tsla.head(21)
```

```python
#passa da False a True, scrive il prezzo col segno negativo (simula un acquisto)
#populates the SimpleAvgPosition column
def shortlongsimple(df):
    for i in range(1,len(df)):
        if (tsla.iloc[i,7] == True) & (tsla.iloc[i-1,7] == False):
            tsla.iloc[i,8] = -1*tsla.iloc[i,3]   #mette il closevalue nella colonna simpleavgposition
        elif (tsla.iloc[i,7] == False) & (tsla.iloc[i-1,7] == True):
            tsla.iloc[i,8] = tsla.iloc[i,3]
        else:
            tsla.iloc[i,8] = 0
```

```python
shortlongsimple(tsla)
```

```python
tsla.head(21)
```

```python
tsla.iloc[0,8] = 0
```

```python
tsla.head(21)
```

```python
# 5. Controlla che se la nuova colonna inizia con una vendita allora finisca con un acquisto, mentre se inizia con un 
#funzione per guardare e capire quante posizione aperte sono long e quante sono short
def countShort(column):
    count = 0
    for i in range(len(tsla)):
        if (tsla.iloc[i,column]< 0):
            count = count+1
    return count

def countLong(column):
    count = 0
    for i in range(len(tsla)):
        if (tsla.iloc[i,column]> 0):
            count = count+1
    return count
```

```python
# 5. Controlla che se la nuova colonna inizia con una vendita allora finisca con un acquisto, mentre se inizia con un 
#acquisto allora finiosca con una vendita. Se non è così, metti a 0 l'ultimo valore diverso da 0.
print(countShort(8))
print(countLong(8))

#Comentario: le posizione aperte e chiuse sono esattamente nella stessa quantita, quindi non bisogna cambiare niente nel DF
```

```python
#6. Calcola la somma di quest'ultima colonna per vedere se la tua attività di trading è stata proficua oppure no.
# Trading basato sulla media mobile semplice non è stato proficuo
tradingSumSimpleAvg = tsla["SimpleAvgPosition"].sum()
print(tradingSumSimpleAvg)
```

```python
# 7. Rifai i punti 2-6 per la media mobile esponenziale. - punto 4
def shortlongexp(df):
    for i in range(1,len(df)):
        if (tsla.iloc[i,10] == True) & (tsla.iloc[i-1,10] == False):
            tsla.iloc[i,11] = -1*tsla.iloc[i,0]
            #print("false2tru")
        elif (tsla.iloc[i,10] == False) & (tsla.iloc[i-1,10] == True):
            tsla.iloc[i,11] = tsla.iloc[i,0]
            #print("true2fals")
        else:
            #print("0")
            tsla.iloc[i,11] = 0
```
