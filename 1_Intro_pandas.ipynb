Importa il file csv __22UsingPandas_EXE.csv__ che contiene i prezzi di Apple e Facebook negli anni 2015-2018. Converti la colonna __Date__ in datetime e mettila come indice del DataFrame. Salva il DataFrame come **stocks**.

```python
import pandas as pd
stocks = pd.read_csv("22UsingPandas_EXE.csv")
stocks  
```

```python
#stocks.head()
stocks = stocks.set_index(pd.to_datetime(stocks.Date)).drop("Date", axis=1)
```

Quanti casi contiene **stocks**? Qual è il prezzo massimo di Facebook in quel periodo? Qual è il prezzo di Apple il 13 maggio 2016? 

```python
print(len(stocks.AAPL))
print(stocks.FB.max())
print(stocks.AAPL.loc["2016-05-13"])
```

Calcola la media dei prezzi per entrambe le azioni a settembre 2017 (due numeri soli devono essere il risultato)

```python
print(stocks.AAPL.loc["2017-09-01":"2017-09-30"].mean())
print(stocks.FB.loc["2017-09-01":"2017-09-30"].mean())
```

Cambia la frequenza a mensile e mostra l'ultimo prezzo del mese. 

```python
stocks.resample("M").first().loc["2017-09-01":"2017-09-30"]
```

Cambia la frequenza a settimanale e mostra il prezzo del venerdì.

```python
#ma quale venerdi?
stocks.resample("W-FRI").last().loc["2017-09-01":"2017-09-30"]
```
