# Calcola il simple return medio annuo usando quello settimanale e la sua volatility annua (su base settimanale) e inseriscili in un dataframe. Fai lo stesso con il log return medio annuo.

```python
#SRannuodaSetti, VolaAnnuadSetti,
#The train of thoughts:
# 1. fare downsampling per settimanale + selezionare solo adj close
# 2. fare weekly simple return
# 3. guardare e scoprire se ci sono settimane col valore NA e fare qualcosa con questi NA
# 4. SRAnnual:  ((SRsettimanaleMedio+1)^numeroDiSettimane)-1 
# 5. Volatilita annuale su ritorno semplice: volansemp=np.sqrt(simpleReturns.var()*250)
# 6. Volatilita annuale su ritorno logaritmico 
```

```python
# 1. fare downsampling per settimanale + selezionare solo adj close
weekDFadjclose=stocks.resample("W").first().loc[:, "Adj Close"]
```

```python
weekDFadjclose
```

```python
# 2. fare weekly simple return
SRweek= (weekDFadjclose - weekDFadjclose.shift(1))/weekDFadjclose.shift(1)
```

```python
SRweek
```

```python
# 3. guardare e scoprire se ci sono settimane col valore NA e fare qualcosa con questi NA
isNAN = SRweek.isnull()
NANrows = isNAN.any(axis=1)
selectNAN = SRweek[NANrows]
print(selectNAN)
```

```python
#solo uno NA, decizione: cancellarlo
SRweek=SRweek.dropna()
SRweek
```

```python
# Per SRannuale e volatilita annuale semplice sara piu facile definire una funzione per ogni uno
# 4. SRAnnuale
# 5. Volatilita
def SR_annual(year):
    year = str(year)
    DF = SRweek.loc[year]
    sr = ((1+DF.mean())**51)-1
    return(sr)
```

```python
SRTable=pd.DataFrame({"SR16":SR_annual(2016), "SR17":SR_annual(2017), "SR18":SR_annual(2018)},tickers)
```

```python
SRTable
```

```python
SRTable = SRTable.transpose()
SRTable
```

```python
SRMean = SRTable.mean()
SRMean
```

```python
volatAnSimp=np.sqrt(SRTable.var()*51)
print(volatAnSimp)
```

```python
def log_year(year):
    year = str(year)
    logWeek = np.log(weekDFadjclose.loc[year])-np.log(weekDFadjclose.loc[year].shift(1))
    logWeek=logWeek.dropna()
    logYear=(logWeek.mean()*(len(logWeek)))
    return(logYear)
```

```python
logYearTable = pd.DataFrame({"LogYear16":log_year(2016),"LogYear17":log_year(2017), "LogYear18":log_year(2018)},tickers)
```

```python
logYearTable
```

```python
logYearTable = logYearTable.transpose()
logYearTable
```

```python
LRMean = logYearTable.mean()
LRMean
```

```python
# la volatility annua su base settimanale partendo dal log return annuale
volatAnLog=np.sqrt(logYearTable.var()*51)
print(volatAnLog)
```

```python
returnVolatTableSimple = pd.DataFrame({"Return": SRMean, "Volatility": volatAnSimp})
returnVolatTableSimple
```

```python
returnVolatTableLog = pd.DataFrame({"Return": LRMean, "Volatility": volatAnLog})
returnVolatTableLog
```
